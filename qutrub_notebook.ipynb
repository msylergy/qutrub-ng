## Basic Verb Conjugation

Let's start with basic verb conjugation using the `conjugate` function.

The main function signature is:
```python
conjugate(verb, future_type, ...)
```

- `verb`: Arabic verb in vocalized form
- `future_type`: For triliteral verbs, the mark of 'Ain' in the future (ÙØªØ­Ø©ØŒ Ø¶Ù…Ø©ØŒ ÙƒØ³Ø±Ø©)
- Additional parameters for specific tenses and forms

````python
# Example 1: Simple conjugation
verb = u"Ø³Ø¹Ø¯"  # "to be happy"
future_type = u"ÙƒØ³Ø±Ø©"  # kasra for the 'ain

# Conjugate in all tenses
result = libqutrub.conjugator.conjugate(
    verb, 
    future_type, 
    alltense=True,
    transitive=True
)

print(f"Conjugation for verb '{verb}':")
print(result)
````

## Different Display Formats

Qutrub supports multiple output formats. Let's explore some of them:

````python
# Test different display formats
verb = u"ÙƒÙØªÙØ¨Ù"  # "to write"
future_type = u"Ø¶Ù…Ø©"  # damma for the 'ain

formats = ["TEXT", "HTML", "DICT", "CSV", "TABLE"]

for fmt in formats:
    print(f"\n=== Format: {fmt} ===")
    try:
        result = libqutrub.conjugator.conjugate(
            verb, 
            future_type, 
            alltense=True, 
            transitive=True,
            display_format=fmt
        )
        if fmt == "DICT":
            print(f"Dictionary with {len(result)} entries")
            # Show first few entries
            first_items = list(result.items())[:3]
            for key, value in first_items:
                print(f"  {key}: {value}")
            if len(result) > 3:
                print(f"  ... and {len(result) - 3} more entries")
        elif fmt == "TABLE":
            print(result[:200] + "..." if len(result) > 200 else result)
        else:
            print(result[:300] + "..." if len(result) > 300 else result)
    except Exception as e:
        print(f"Error with format {fmt}: {e}")
````

## Verb Validation

Before conjugating, we should validate if the verb is a proper infinitive form.

````python
# Test verb validation
test_verbs = [u"ÙƒÙØªÙØ¨Ù", u"Ø¶ÙØ±ÙØ¨Ù", u"Ø³ÙÙƒÙØªÙ", u"ÙƒØªØ¨", u"invalidverb"]

for verb in test_verbs:
    is_valid = libqutrub.verb_valid.is_valid_infinitive_verb(verb)
    print(f"Verb '{verb}': {'Valid' if is_valid else 'Invalid'}")
    
    if is_valid:
        # Get potential future types
        future_types = get_future_type_entree(verb)
        print(f"  Possible future types: {future_types}")
````

## Specific Tense Conjugation

Let's conjugate specific tenses instead of all tenses.

````python
# Conjugate specific tenses
verb = u"Ø°ÙÙ‡ÙØ¨Ù"  # "to go"
future_type = u"ÙÙØªÙ’Ø­Ø©"  # fatha for the 'ain

tenses = [
    ("Past", lambda: libqutrub.conjugator.conjugate(verb, future_type, past=True, transitive=True)),
    ("Future/Present", lambda: libqutrub.conjugator.conjugate(verb, future_type, future=True, transitive=True)),
    ("Imperative", lambda: libqutrub.conjugator.conjugate(verb, future_type, imperative=True, transitive=True)),
    ("Passive", lambda: libqutrub.conjugator.conjugate(verb, future_type, passive=True, transitive=True))
]

for tense_name, conj_func in tenses:
    print(f"\n=== {tense_name} Tense ===")
    try:
        result = conj_func()
        print(result[:400] + "..." if len(result) > 400 else result)
    except Exception as e:
        print(f"Error: {e}")
````

## Working with Different Verb Types

Let's test various verb types, including transitive and intransitive verbs.

````python
# Test different verb types
verbs_data = [
    (u"ÙƒÙØªÙØ¨Ù", u"Ø¶Ù…Ø©", "transitive", "to write"),
    (u"Ø¶ÙØ±ÙØ¨Ù", u"ÙƒØ³Ø±Ø©", "transitive", "to hit"),
    (u"Ø³ÙÙƒÙØªÙ", u"Ø¶Ù…Ø©", "intransitive", "to be silent"),
    (u"Ø³ÙÙƒÙÙ†Ù", u"Ø¶Ù…Ø©", "transitive", "to make people calm"),
    (u"Ø¹ÙÙ„Ù‘ÙÙ…Ù", u"," "transitive", "to teach"),
    (u"Ø£ÙÙƒÙ’Ø±ÙÙ…Ù", u"," "transitive", "to honor"),
    (u"ØªÙÙÙÙ‚Ù‘ÙØ¯Ù", u"," "transitive", "to check"),
    (u"ØµÙØ§Ø±ÙØ¹Ù", u"," "transitive", "to wrestle"),
    (u"Ø§ÙØ¯Ù‘ÙØ§Ø±ÙÙƒÙ", u"," "transitive", "to realize"),
]

for verb, future_type, trans_type, meaning in verbs_data:
    print(f"\n=== {verb} ({meaning}) ===")
    
    # Check if valid
    is_valid = libqutrub.verb_valid.is_valid_infinitive_verb(verb)
    if not is_valid:
        print(f"  Invalid infinitive verb")
        continue
    
    # Set transitivity
    is_transitive = trans_type == "transitive"
    
    try:
        # Conjugate with dictionary format for cleaner output
        result = libqutrub.conjugator.conjugate(
            verb, 
            future_type if future_type != "," else u"ÙƒØ³Ø±Ø©",  # Default to kasra if not specified
            alltense=False,
            past=True,  # Show only past tense for brevity
            transitive=is_transitive,
            display_format="COMPACT"  # Try compact format
        )
        
        if isinstance(result, dict):
            # Show a few keys from the dictionary
            items = list(result.items())[:5]
            for key, value in items:
                print(f"  {key}: {value}")
            if len(result) > 5:
                print(f"  ... and {len(result) - 5} more entries")
        else:
            # Limit output length for readability
            print(result[:200] + "..." if len(result) > 200 else result)
            
    except Exception as e:
        print(f"  Error conjugating: {e}")
````

## Advanced Features

Let's explore some advanced features like confirmed forms and mood variations.

````python
# Advanced conjugation features
verb = u"ÙƒÙØªÙØ¨Ù"
future_type = u"Ø¶Ù…Ø©"

advanced_options = [
    ("Future Subjunctive (Ù…ÙÙ†Ù’ØµÙÙˆØ¨)", lambda: libqutrub.conjugator.conjugate(
        verb, future_type, future=True, future_moode=True, transitive=True, display_format="TEXT")),
    ("Future Jussive (Ù…ÙØ¬Ù’Ø²ÙÙˆÙ…)", lambda: libqutrub.conjugator.conjugate(
        verb, future_type, future=True, future_moode=False, transitive=True, display_format="TEXT")),
    ("Confirmed Future", lambda: libqutrub.conjugator.conjugate(
        verb, future_type, future=True, confirmed=True, transitive=True, display_format="TEXT"))
]

for name, conj_func in advanced_options:
    print(f"\n=== {name} ===")
    try:
        result = conj_func()
        print(result[:300] + "..." if len(result) > 300 else result)
    except Exception as e:
        print(f"Error: {e}")
````
